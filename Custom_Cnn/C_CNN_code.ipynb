```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_data_rules = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest'
)

val_data_rules = ImageDataGenerator(rescale=1./255)
test_data_rules = ImageDataGenerator(rescale=1./255)
```

```python
labels = ['Baked Potato', 'Crispy Chicken', 'Donut', 'Fries', 'Hot Dog', 'Sandwich', 'Taco', 'Taquito', 'apple_pie', 'burger', 'butter_naan', 'chai', 'chapati', 'cheesecake', 'chicken_curry', 'chole_bhature', 'dal_makhani', 'dhokla', 'fried_rice', 'ice_cream', 'idli', 'jalebi', 'kaathi_rolls', 'kadai_paneer', 'kulfi', 'masala_dosa', 'momos', 'omelette', 'paani_puri', 'pakode', 'pav_bhaji', 'pizza', 'samosa', 'sushi']
```

```python
final_train_data = train_data_rules.flow_from_directory(
    training_data_path,
    color_mode='rgb',
    target_size=(256, 256),
    class_mode='categorical',
    batch_size=60,
    shuffle=True
)

final_val_data = val_data_rules.flow_from_directory(
    validation_data_path,
    color_mode='rgb',
    target_size=(256, 256),
    class_mode='categorical',
    shuffle=False
)

final_test_data = test_data_rules.flow_from_directory(
    testing_data_path,
    color_mode='rgb',
    target_size=(256, 256),
    class_mode='categorical',
    shuffle=False
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense,Activation
from tensorflow.keras.activations import relu,sigmoid,softmax
from tensorflow.keras.models import load_model
```

```python
model = Sequential()

model.add(Conv2D(
    128, (3,3),
    activation='relu',
    kernel_initializer='he_uniform',
    strides=1,
    padding='same',
    input_shape=(256,256,3)
))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3), activation='relu', kernel_initializer='he_uniform'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), activation='relu', kernel_initializer='he_uniform'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(16, (3,3), activation='relu', kernel_initializer='he_uniform'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(4, (3,3), activation='relu', kernel_initializer='he_uniform'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())  # ← parentheses are IMPORTANT
```

```python
model.add(Dense(units = 128,
                activation='relu',
                kernel_initializer='he_uniform'))
model.add(Dense(units = 64,
                activation='relu',
                kernel_initializer='he_uniform'))
model.add(Dense(units = 32,
                activation='relu',
                kernel_initializer='he_uniform'))
model.add(Dense(units = 16,
                activation='relu',
                kernel_initializer='he_uniform'))
model.add(Dense(units = 4,
                activation='relu',
                kernel_initializer='he_uniform'))

model.add(Dense(units=34,
                activation='softmax',
                kernel_initializer='glorot_uniform'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam(learning_rate=1e-5)',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(final_train_data,
          epochs=10,
          validation_data=final_val_data)
```

```python
train_loss, train_acc = model.evaluate(final_train_data, verbose=0)
val_loss, val_acc = model.evaluate(final_val_data, verbose=0)
test_loss, test_acc = model.evaluate(final_test_data, verbose=0)

print("Train:", train_acc)
print("Val:", val_acc)
print("Test:", test_acc)
```

```python
import matplotlib.pyplot as plt

epochs = range(1, len(history.history['accuracy'])+1)

plt.figure(figsize=(10,4))

# Accuracy
plt.subplot(1,2,1)
plt.plot(epochs, history.history['accuracy'], label='Train Acc')
plt.plot(epochs, history.history['val_accuracy'], label='Val Acc')
plt.title("Accuracy")
plt.legend()

# Loss
plt.subplot(1,2,2)
plt.plot(epochs, history.history['loss'], label='Train Loss')
plt.plot(epochs, history.history['val_loss'], label='Val Loss')
plt.title("Loss")
plt.legend()

plt.show()
```

```python
import numpy as np

# Predict probabilities for all test images
y_pred_probs = model.predict(final_test_data, verbose=1)

# Convert probabilities → class index
y_pred = np.argmax(y_pred_probs, axis=1)

# True labels directly from generator
y_true = final_test_data.classes

# Class names
labels = list(final_test_data.class_indices.keys())

print("Prediction completed ✅")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# ----- SHOW REPORT ON SCREEN -----
print("\n===== CLASSIFICATION REPORT =====\n")
print(classification_report(y_true, y_pred, target_names=labels))

# ----- STORE REPORT FOR JSON -----
report = classification_report(
    y_true,
    y_pred,
    target_names=labels,
    output_dict=True
)

# ----- CONFUSION MATRIX -----
cm = confusion_matrix(y_true, y_pred)

print("\n===== CONFUSION MATRIX =====\n")
print(cm)
```

```python
import json

# safe history extraction (prevents crash if history missing)
train_acc_epoch = history.history['accuracy'] if 'accuracy' in history.history else []
val_acc_epoch   = history.history['val_accuracy'] if 'val_accuracy' in history.history else []
train_loss_epoch = history.history['loss'] if 'loss' in history.history else []
val_loss_epoch   = history.history['val_loss'] if 'val_loss' in history.history else []

results = {
    "train_accuracy": float(train_acc),
    "validation_accuracy": float(val_acc),
    "test_accuracy": float(test_acc),

    "train_accuracy_per_epoch": train_acc_epoch,
    "val_accuracy_per_epoch": val_acc_epoch,
    "train_loss_per_epoch": train_loss_epoch,
    "val_loss_per_epoch": val_loss_epoch,

    "classification_report": report
}

with open("/kaggle/working/training_results.json", "w") as f:
    json.dump(results, f, indent=4)

print("training_results.json saved ✅")
```
